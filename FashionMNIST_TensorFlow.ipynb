```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

print(tf.config.list_physical_devices('GPU'))

(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']



x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = (x_train - 0.5) / 0.5
x_test = (x_test - 0.5) / 0.5

x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

y_train_categorical = keras.utils.to_categorical(y_train, 10)
y_test_categorical = keras.utils.to_categorical(y_test, 10)

def show_samples():
    fig, axes = plt.subplots(2, 5, figsize=(12, 6))

    for i in range(10):
        img = x_train[i].squeeze()
        label = y_train[i]

        row = i // 5
        col = i % 5

        axes[row, col].imshow(img, cmap='gray')
        axes[row, col].set_title(f'{class_names[label]}')
        axes[row, col].axis('off')

    plt.tight_layout()
    plt.show()

show_samples()

#Model architecture
def create_fashion_cnn():
    model = keras.Sequential([

        layers.Conv2D(32, (3, 3), padding='same', input_shape=(28, 28, 1)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(64, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(128, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Flatten(),

        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])

    return model

model = create_fashion_cnn()
print(model.summary())

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

class TrainingCallback(keras.callbacks.Callback):
    def __init__(self):
        self.batch_count = 0

    def on_batch_end(self, batch, logs=None):
        self.batch_count += 1
        if self.batch_count % 200 == 0:
            epoch = self.batch_count // (len(x_train) // 64)
            print(f'Epoch [{epoch+1}], Step [{self.batch_count % (len(x_train) // 64)}/{len(x_train) // 64}], '
                  f'Loss: {logs["loss"]:.4f}')

batch_size = 64
num_epochs = 10

history = model.fit(
    x_train, y_train_categorical,
    batch_size=batch_size,
    epochs=num_epochs,
    validation_split=0.0,
    callbacks=[TrainingCallback()],
    verbose=1
)

def test_model(model, x_test, y_test):
    predictions = model.predict(x_test)
    predicted_classes = np.argmax(predictions, axis=1)

    test_loss, test_accuracy = model.evaluate(x_test, y_test_categorical, verbose=0)
    overall_accuracy = test_accuracy * 100
    print(f'Overall ACC on test set: {overall_accuracy:.2f}%')

    class_correct = [0 for _ in range(10)]
    class_total = [0 for _ in range(10)]

    for i in range(len(y_test)):
        label = y_test[i]
        class_total[label] += 1
        if predicted_classes[i] == label:
            class_correct[label] += 1

    print('\nACC Based on class:')
    for i in range(10):
        if class_total[i] > 0:
            accuracy = 100 * class_correct[i] / class_total[i]
            print(f'{class_names[i]}: {accuracy:.2f}%')

    return overall_accuracy

test_accuracy = test_model(model, x_test, y_test)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.subplot(1, 2, 2)
plt.plot([acc * 100 for acc in history.history['accuracy']])
plt.title('Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')

plt.tight_layout()
plt.show()

def show_misclassified_samples(model, x_test, y_test, num_samples=10):
    predictions = model.predict(x_test)
    predicted_classes = np.argmax(predictions, axis=1)

    misclassified_indices = []
    for i in range(len(y_test)):
        if predicted_classes[i] != y_test[i] and len(misclassified_indices) < num_samples:
            misclassified_indices.append(i)

    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    for i, idx in enumerate(misclassified_indices):
        row = i // 5
        col = i % 5

        img = x_test[idx].squeeze()
        actual = y_test[idx]
        predicted = predicted_classes[idx]

        axes[row, col].imshow(img, cmap='gray')
        axes[row, col].set_title(f'Actual: {class_names[actual]}\n'
                                f'Predicted: {class_names[predicted]}')
        axes[row, col].axis('off')

    plt.tight_layout()
    plt.show()

print("\nWrong output:")
show_misclassified_samples(model, x_test, y_test)
```
